## Transformers: Operators algebra, some similarity with Quantum Fields Theory 

The transition from Classical Mechanics to Quantum Mechanics is reflected in transition from "numbers" to operators. 
In recent years , we may see the same in Natural language processing. 
In last NLP developments,  words are no longer represented by numbers, but represented by vectors, for example as 512-D embedding in some space. It force us to work with operators, not numbers.

The biggest manipulator with vectors in Quantum Fields Theory (and Quntum mechanics as its part). The NLP is very different in nature, but we can try to borrow some methods and ideas of manipulation with a vectors. 
We may try to think in active terms : we come from one place to another with some probability or we consume some state and produce another state.
